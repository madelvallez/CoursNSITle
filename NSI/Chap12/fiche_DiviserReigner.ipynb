#Méthode Diviser pour Reigner

##I. Definition

La méthode *diviser pour reigner* est une méthode basée sur le principe de diviser une grosse tache en plusieurs plus petites taches. On les résoud puis on les réassemble pour obtenir le résultat final. 

On retrouve trois étapes:
* Diviser: on découpe le problème d'origine en sous-problèmes
* Reigner: on résout les sous-problèmes
* Fusionner: on fusionnne les solutions aux sous-problèmes pour obtenir la solution au problème d'origine.

##II. Recherche d'un minimum dans une liste

####Rappel: 
* `tableau[:i]` renvoie la tranche du tableau entre `tableau[0] `et `tableau[i-1]`.
* `tableau[i:]` renvoie la tranche du tableau entre `tableau[i]` et `tableau[len(t)-1]`

````
def minimum(liste):
    taille = len(liste)
    if taille == 1 :
        return liste[0]
    else :
        milieu = taille // 2
        x = minimum(liste[:milieu])
        y = minimum(liste[milieu:])
        if x < y :
            return x
        else :
            return y
````

Dans cet exemple, 
```
milieu = taille // 2
x = minimum(liste[:milieu])
y = minimum(liste[milieu:])
```
correspond à la partie *diviser*,

`return liste[0]`

correspond à la partie *reigner*,
```
if x < y :
    return x
else :
    return y
```
correspond à la partie *fusionner*.

####Remarque: 
Dans cette situation, la méthode *diviser pour reigner* ne permet pas de rédure la complexité de l'algorithme qui est **linéaire**.

###2) Diviser dans `tri_fusion()`

````
def division(liste):
    taille = len(liste)    
    milieu = taille // 2
    return liste[:milieu], liste[milieu:]

print(division([3,4,6,2,5,1,8,7,9]))
````

###3) Reigner dans `tri_fusion`

À force de diviser la liste, on retrouve des listes d'un seul élément -> elle est triée

###4) Fusionner avec `tri_fusion`

````
def fusion(gauche,droite):
    if   gauche == []          : return droite
    elif droite == []          : return gauche
    elif gauche[0] < droite[0] : return [gauche[0]] + fusion(gauche[1:],droite)
    else                       : return [droite[0]] + fusion(gauche,droite[1:])
````

###5) La fonction `tri_fusion`

````
def tri_fusion(liste):
    if len(liste) <= 1 :
        return liste
    else :
        gauche, droite = division(liste)
        return fusion (tri_fusion(gauche), tri_fusion(droite))
````
